**One Hot Encoding** with a column

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
## Creating a color columns to encode
df=pd.DataFrame({
    'color':['red','blue','blue','black','green']
    })
```

```python
df
```

```python
## Initialze onehotencoder
encoder=OneHotEncoder()
```

```python
## Fit and transfrom encoded color column data
encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
encoded_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out(['color']))
```

```python
encoded_df
```
